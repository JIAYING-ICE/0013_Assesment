Step 01: Dowloading Data to Local Computer

```python
import os
import urllib.request
import pandas as pd
import geopandas as gpd

# 1. Ensure data folder exists by creating a function
def ensure_data_folder():
    if not os.path.isdir("data"):
        os.makedirs("data")
        print("Created folder: data/")
    else:
        print("Folder already exists: data/")
    print("—" * 80) # continues line

ensure_data_folder() # Run Function

# 2. Helper: download only if missing + confirmations
def download_if_missing(url, local_name):
    dest = os.path.join("data", local_name)
    
    print(f" let me check: {local_name}")

    if os.path.isfile(dest):
        size = round(os.path.getsize(dest) / 1024 / 1024, 2)
        print(f"File already exists ({size} MB) — skipping download.")
    else:
        print(f"Downloading from:\n     {url}")
        urllib.request.urlretrieve(url, dest)

        size = round(os.path.getsize(dest) / 1024 / 1024, 2)
        print(f"Download complete — saved {size} MB")

    print("—" * 80)
    return dest
```

**Step 02:** Base file from the UCL Orca and London Goverment

```python
base = "https://orca.casa.ucl.ac.uk/~jreades/data/"

orca_files = [
    "20250615-London-reviews.csv.gz",
    "20250615-London-listings.csv.gz",
    "MSOA-2011.gpkg"
]

# add London DataStore direct CSV links
extra_files = {
    "Persons_per_dwelling.csv":
    "https://data.london.gov.uk/download/2n8gn/663f3592-8358-4a10-bff2-7b45a758f362/Persons_per_dwelling.csv",

    "local-authority-rents-borough.xlsx":
    "https://data.london.gov.uk/download/2g1k1/2a74e893-3188-4cb8-b5a7-087a1965b623/local-authority-rents-borough.xlsx",

    "london_boroughs.geojson":
    "https://opendata.arcgis.com/datasets/c6fbef8d68424c5cbcdf842c16ed4562_48.geojson"
}

local_paths = []

print("!STARTING ORCA DOWNLOADS!")

# Download ORCA DATA
for fname in orca_files:
    url = base + fname
    path = download_if_missing(url, fname)
    local_paths.append(path)

# Download DATASTORE DATA
for local_name, url in extra_files.items():
    path = download_if_missing(url, local_name)
    local_paths.append(path)
```

Step 03: Transforming file from Excel to CSV

```python
# Convert the Excel rents file to CSV
excel_path = "data/local-authority-rents-borough.xlsx"
csv_path   = "data/local-authority-rents-borough.csv"

# Read Excel
rents_xlsx = pd.read_excel(excel_path)

# Save as CSV
rents_xlsx.to_csv(csv_path, index=False)

print("✔ Converted Excel to CSV:", csv_path)
```

Step 04: Reading the file into pandas and geopandas

```python
print("!LOADING FILES!")

# Reviews CSV.GZ
reviews = pd.read_csv("data/20250615-London-reviews.csv.gz", compression="gzip", low_memory=False)
print(f"✔ reviews loaded: {reviews.shape[0]:,} rows, {reviews.shape[1]} columns")

# Listings CSV.GZ
listings = pd.read_csv("data/20250615-London-listings.csv.gz", compression="gzip", low_memory=False)
print(f"✔ listings loaded: {listings.shape[0]:,} rows, {listings.shape[1]} columns")

# MSOA GPKG
msoa = gpd.read_file("data/MSOA-2011.gpkg")
print(f"✔ msoa loaded: {msoa.shape[0]:,} polygons, {msoa.shape[1]} columns")

# Rent per borough EXCEL
rent = pd.read_csv("data/local-authority-rents-borough.csv", low_memory=False)
print(f"✔ rent loaded: {rent.shape[0]:,} rows, {rent.shape[1]} columns")

# Dwelling CSV
dwelling = pd.read_csv("data/Persons_per_dwelling.csv", low_memory=False)
print(f"✔ dwelling loaded: {dwelling.shape[0]:,} rows, {dwelling.shape[1]} columns")

# Borough GSON
borough = gpd.read_file("data/london_boroughs.geojson")
print(f"✔ borough loaded: {borough.shape[0]:,} polygons, {borough.shape[1]} columns")



print("**All files successfully loaded.**")
```

Step 05: Clean Columns 

```python
def clean_names(df):
    df = df.copy()
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
        .str.replace(r"[^\w]", "", regex=True)
    )
    return df

# put all your loaded dataframes into a dict
dfs = {
    "reviews": reviews,
    "listings": listings,
    "msoa": msoa,
    "rent": rent,
    "dwelling": dwelling,
    "borough": borough
}

# clean each dataframe
for file, frame in dfs.items():
    dfs[file] = clean_names(frame)
    print(f" cleaned column names for: {file}")

# extract back if you want individual variables
reviews   = dfs["reviews"]
listings  = dfs["listings"]
msoa      = dfs["msoa"]
rent     = dfs["rent"]
dwelling  = dfs["dwelling"]
borough   = dfs["borough"]
```
